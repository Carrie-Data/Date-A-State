```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/Raw_data_from_Kaggle.csv"

# Read Purchasing File and store into Pandas data frame
Life_data = pd.read_csv(file_to_load, encoding = 'ISO-8859-1')
```

```python
Life_data.head(2)
```

```python
del Life_data['UID']
del Life_data['BLOCKID']
del Life_data['SUMLEVEL']
del Life_data['female_age_stdev']
del Life_data['female_age_sample_weight']
del Life_data['male_age_stdev']
del Life_data['male_age_sample_weight']
del Life_data['hc_mortgage_stdev']
del Life_data['hc_mortgage_sample_weight']
del Life_data['hc_mortgage_samples']
del Life_data['home_equity_second_mortgage']
del Life_data['second_mortgage']
del Life_data['second_mortgage_cdf']
del Life_data['primary']
del Life_data['zip_code']
del Life_data['area_code']
del Life_data['rent_stdev']
del Life_data['rent_sample_weight']
del Life_data['rent_samples']
del Life_data['universe_samples']
del Life_data['used_samples']
del Life_data['family_median']
del Life_data['family_stdev']
del Life_data['family_sample_weight']
del Life_data['hi_sample_weight']
del Life_data['hi_samples']
del Life_data['hi_stdev']
del Life_data['family_samples']
del Life_data['hc_stdev']
del Life_data['hc_samples']
del Life_data['family_mean']
del Life_data['hs_degree']
del Life_data['hs_degree_male']
del Life_data['hs_degree_female']
del Life_data['female_age_samples']
del Life_data['COUNTYID']
del Life_data['STATEID']
del Life_data['hc_sample_weight']
del Life_data['hc_mortgage_mean']
del Life_data['hc_mortgage_median']
del Life_data['male_age_samples']
```

```python
Life_data.columns
```

```python
Life_data.head()
```

```python
Life_data['single'] = 1 - Life_data['married'] - Life_data['married_snp'] - Life_data['separated'] - Life_data['divorced']
```

```python
Life_data.head()
```

```python
Life_data['married'] = Life_data['married'] + Life_data['married_snp']
del Life_data['married_snp']
Life_data
```

```python
Life_data = Life_data.dropna(how='any')
```

```python
By_State_life_data = Life_data.groupby('state')
```

```python
income = By_State_life_data["hi_mean"].mean()
land = By_State_life_data["ALand"].sum()
water = By_State_life_data["AWater"].sum()
rent = By_State_life_data['rent_mean'].mean()
mortgage = By_State_life_data['hc_mean'].mean()
population = By_State_life_data['pop'].sum()
female_pop = By_State_life_data['female_pop'].sum()
male_pop = By_State_life_data['male_pop'].sum()
single = By_State_life_data['single'].mean()
married = By_State_life_data['married'].mean()
seperated = By_State_life_data['separated'].mean()
divorced = By_State_life_data['divorced'].mean()
```

```python
aster_dataframe = pd.DataFrame({'income':income,
                               'land':land,
                               'water':water,
                               'rent': rent,
                               'mortgage': mortgage,
                               'population': population,
                               'female_pop': female_pop,
                               'male_pop': male_pop,
                               'single': single,
                               'married': married,
                               'seperated': seperated,
                               'divorced': divorced})
aster_dataframe.head()
```

```python
aster_dataframe['land_percentage'] =aster_dataframe['land'] /(aster_dataframe['land'] + aster_dataframe['water']) 
aster_dataframe['water_percentage'] =aster_dataframe['water'] /(aster_dataframe['land'] + aster_dataframe['water']) 
aster_dataframe.head()
```

```python
del aster_dataframe['land']
del aster_dataframe['water']
```

```python
Life_data.to_csv('./Resources/Life_data_full_cleaned_dataset.csv')
aster_dataframe.to_csv('./Resources/Life_data_agg_aster_plot.csv')
```
